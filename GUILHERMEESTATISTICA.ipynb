```python
altura_media = dados['Renda'].mean()
altura_media
```

# Relação da variável Renda

```python
if (altura_moda<altura_mediana<altura_media):
    print("Asimetria à direita")
elif(altura_moda<altura_mediana<altura_media):
    print("Assimetria á esquerda")
elif(altura_moda==altura_mediana==altura_media):
    print("É uma relação simétrica")
else:
    print("Não foi possível analisar a simetria")
```

# Combinações

![title](combinacao.jpg)

```python
from scipy.special import comb
```

```python
# No jogo da megasena tem 60 numeros para marcar , porem o jogo minimo é de 6 numeros
# Entao queremos saber quatnas combinações de 6 numeros podem ser formadas com os 60 numeros disponíveis

combinacoes = comb(60,6)
combinacoes
```

```python
# Probabilidade de com um só jogo ganahr a mega sena
probabilidade = 1/combinacoes
probabilidade

print("******* A chance de ganhar na megasena com 1 só jogo é de: ", ('%0.15f' % probabilidade) )
```
